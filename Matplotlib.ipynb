```python
#l is x-axis and v is y-axis
l=['kajal','parvani','kazi','rida','mukesh','philip','bhuvneshwari','nikhil','sagar','anandha']
v=np.random.randint(50,100,10)

colors=['red','yellow','orange','blue','green','pink','hotpink','brown','purple']

ax=plt.axes()
ax.set_facecolor('yellow')

plt.bar(l,v,width=0.3,align='edge',edgecolor='red',color=colors)
plt.xticks(rotation=45)
plt.xlabel('Learner')
plt.ylabel('Marks')
plt.title('Assessment Marks')
plt.show()
```

```python
#align sai jo bar hai wo left side mai chale jyaga
l=['kajal','parvani','kazi','rida','mukesh','philip','bhuvneshwari','nikhil','sagar','anandha']
v=np.random.randint(50,100,10)

colors=['red','yellow','orange','blue','green','pink','hotpink','brown','purple']

ax=plt.axes()
ax.set_facecolor('yellow')

plt.barh(l,v,height=0.3,align='edge',edgecolor='red',color=colors)
plt.xticks(rotation=45)
plt.xlabel('Learner')
plt.ylabel('Marks')
plt.title('Assessment Marks')
plt.show()
```

```python
city=['delhi','mumbai','chennai','amritsar','hydrabad','banglore']
male=[60,45,55,66,75,58]
female=[30,40,45,55,50,55]
gender=['Male','Female']

ax=plt.axes()
ax.set_facecolor('yellow')

plt.bar(city,male,color='green',edgecolor='black')
plt.bar(city,female,color='red',edgecolor='black',bottom=female)

plt.legend(gender)
plt.show()
```

```python
city=['delhi','mumbai','chennai','amritsar','hydrabad','banglore']
male=[60,45,55,66,75,58]
female=[30,40,45,55,50,55]
gender=['Male','Female']

plt.subplot(1,2,1)
plt.bar(city,male,color='green',edgecolor='black')
plt.xticks(rotation=45)
plt.legend(gender[0])

plt.subplot(1,2,2)
plt.bar(city,female,color='red',edgecolor='black')
plt.xticks(rotation=45)
plt.legend(gender[1])

plt.show()
```

```python
city=['delhi','mumbai','chennai','amritsar','hydrabad','banglore']
male=[60,45,55,66,75,58]
female=[30,40,45,55,50,55]
gender=['Male','Female']

ax=plt.axes()
ax.set_facecolor('yellow')

pop=np.arange(len(city))
#eisme width aur pop+0.3 same hai
plt.bar(pop,male,color='green',width=0.3,edgecolor='black')
plt.bar(pop+0.3,female,color='red',width=0.3,edgecolor='black')
#pop jo hai wo index hai city ki to audhar wo fir city execute hogi
plt.xticks(pop,city)
plt.legend(gender)

plt.show()
```
